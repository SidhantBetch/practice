```python
y = df[["Churn"]]
x = df[["Age","Gender","Tenure","MonthlyCharges"]]
```

```python
x["Gender"] = x["Gender"].apply(lambda x: 1 if x == "Female" else 0)
```

```python
x.head()
```

```python
type(x["Gender"][0])
```

```python
y["Churn"] = y["Churn"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
from sklearn.model_selection import train_test_split
x_tarin, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_tarin)
```

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
def modelperformance(predictions):
    print("Accuracy score on model is {}".format(accuracy_score(y_test, predictions)))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_mode = LogisticRegression()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
log_mode.fit(x_train, y_train)
```

```python
y_pred = log_mode.predict(x_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
param_grid = {
    "n_neighbors" : [3,5,7,9],
    "weights" : ["uniform", "distance"],
}
```

```python
gridkn = GridSearchCV(KNeighborsClassifier(), param_grid, cv = 5)
```

```python
gridkn.fit(x_train, y_train)
```

```python
gridkn.best_params_
```

```python
y_pred = gridkn.predict(x_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
param_grid = {
    "C": [0.01, 0.1, 0.5, 1],
    "kernel": ["linear", "rbf", "poly"],
}
```

```python
gridsvc = GridSearchCV(svm,  param_grid, cv = 5)
```

```python
gridsvc.fit(x_train, y_train)
```

```python
gridsvc.best_params_
```

```python
y_pred = gridsvc.predict(x_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
param_grid = {
    "criterion" : ["gini", "entropy"],
    "splitter" : ["best", "random"],
    "max_depth" : [None, 10, 20, 30],
    "min_samples_split" : [2,5,10],
    "min_samples_leaf" : [1,2,4]
}
```

```python
grid_tree = GridSearchCV(DecisionTreeClassifier(), param_grid, cv = 5)
```

```python
grid_tree.fit(x_train, y_train)
```

```python
grid_tree.best_params_
```

```python
y_pred = grid_tree.predict(x_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier()
```

```python
param_grid = {
    "n_estimators" : [32,64,128,256],
    "max_features" : [2,3,4],
    "bootstrap" : [True, False]
}
```

```python
grid_rfc = GridSearchCV(rfc_model, param_grid, cv = 5)
```

```python
grid_rfc.fit(x_train, y_train)
```

```python
grid_rfc.best_params_
```

```python
y_pred = grid_rfc.predict(x_test)
```

```python
modelperformance(y_pred)
```

```python
best_model = grid_rfc.best_estimator_
```

```python
joblib.dump(best_model, "model.pkl")
```
